# Tipos de datos (I)
- Podemos clasificar tres formatos basicos de los datos conforme a la estructuracion de la informacion:
    - Datos estructurados : Toda la informacion esta identificada conforme a un patron comun y definido que representa un modelo .optima para ser analizada . Ej Modelo relacional, base de datos SQL . Donde tenemos en una base de datos , donte tenemos las tablas en perfecto estado y estan interaccionadas entre ellas y demas .
    - Datos semi- estructurados.: La estructura de los datos esta definida y es conocida, pero no responde a un modelo estricto. Optima para ser procesada  Ej . JSON,XML . Son datos que tienen marcas , es como un etiquetado de datos 
    - Datos no estructurados: No representa un modelo y carece de estructura conocida . Requiere ser trasformada para su analisis . Ej . Multimedia , texto libre . Tenemos que hacer mucho trabajo para pasarlo a semi o completo estructurado 

- Necesitamos tener los dos primeros para trabajar en Machine Learning 
- El problema es que el  80% de internet esta en dato no estructurado .

# Data wrangling: Transformado nuestro datos
- Objetivos: Transformar datos en bruto(raw data) en un formato mas apropiado y valioso desde el punto de vista del analisis.
- El proceso debe realizarse a partir de la forma mas pura posible de los datos en bruto y debe documentarse para ser reproducible.
- No es suficiente con obtener un conjunto de datos estructurado  ya que no siempre es optimo para el analisis. No por que esten estructurados ya sean buenos datos . 
    - Datos estructurados =/= Buenos datos
- Generalmente buscamos que los datos mantengan unas propiedades deseables que conocemos como : Tidy Data. Al concepto de convertir datos brutos a datos que se puedan trabajar en machine learning se llama Tidy Data.
